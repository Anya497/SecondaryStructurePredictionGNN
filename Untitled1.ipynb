```python
import matplotlib.pyplot as plt
import os
from skimage.io import imread
```

```python
fig = plt.figure(figsize=(12, 45))
columns = 3
rows = 60
ax = []
k = 1
for im_name in sorted(os.listdir("./pictures/0/"))[1::]:
    img = imread("./pictures/1/" + im_name)
    # create subplot and append to ax
    ax.append( fig.add_subplot(rows, columns, k)  )
    ax[-1].set_title(im_name)  # set title
    plt.imshow(img)
    k += 1

# do extra plots on selected axes/subplots
# note: index starts with 0
# ax[2].plot(xs, 3*ys)
# ax[19].plot(ys**2, xs)

plt.show()  
```
